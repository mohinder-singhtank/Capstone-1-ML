```python
import mysql.connector as sqcon
import pandas as pd
```

```python
def create_sql_conn():
    # Update the code to use exception handling
    # https://dev.mysql.com/doc/connector-python/en/connector-python-example-connecting.html
    conn = sqcon.connect(user='student', 
            password='student',
            host='cpanel.insaid.co',
            database='Capstone1')
    return conn
```

```python
def getTableData(table_name):
    conn = create_sql_conn()
    db_cursor = conn.cursor()
    db_cursor.execute('SELECT * FROM ' + table_name)
    table_rows = db_cursor.fetchall()
    df = pd.DataFrame(table_rows)
    conn.close()
    return df
```

```python
gender_age_train_df = getTableData('gender_age_train')
gender_age_train_df.columns =['DeviceId', 'Gender', 'Age', 'Age_Group']
phone_brand_device_model_df = getTableData('phone_brand_device_model')
phone_brand_device_model_df.columns =['DeviceId', 'Brand', 'Model_Name']
# Assuming the events_data file os in the folder named data 
events_data_df = pd.read_csv('data/events_data.csv')
```
